*정답 : 2번*

pandas 열 접근 방식
- 딕셔너리 스타일 (df["column_name"]):
- 속성 스타일 (df.column_name)

df.column_name = 0은 dataframe에 열로 추가 되는게 아니라 dataframe의 속성으로 추가됨

```python
import pandas as pd

# 빈 DataFrame 생성
df = pd.DataFrame()

# 1. 속성 추가
df.data_column = 0  # 속성 추가
print("속성 추가 후 df.columns:", df.columns)  # DataFrame의 열 확인 (빈 리스트 출력)
print("속성으로 접근한 df.data_column:", df.data_column)  # 속성 확인

# 2. 열 추가
df["data_column"] = 0  # 실제 열 추가
print("\n열 추가 후 df.columns:", df.columns)  # DataFrame의 열 확인
print("열로 접근한 df['data_column']:\n", df["data_column"])  # 열 확인
```

```python
import pandas as pd
from scipy import stats


# 예제 데이터프레임 생성
data = {'longitude': [100.1, 102.2, 101.5], 'latitude': [200.3, 201.8, 202.4]}
df = pd.DataFrame(data)
df['longitude'].mean()

desc = stats.describe(df)

print(desc.mean)
print(desc.mean[0])
```
