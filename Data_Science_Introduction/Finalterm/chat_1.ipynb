### CSV 파일 특징
 - 장점 : 형식이 단순하여 직관적 처리 가능, 구조화된 데이터를 효율적으로 처리할 수 있음
 - 단점 : 구조 정보를 별도로 저장해야 함(예: 데이터타입, 특정필드 길이 제한, 필수값 등)

 1 => csv은 구분자, 공백, 줄바꿈과 같으 규칙을 유지해야 함으로 txt파일보다 용량이 더 클 수 있음 txt파일은 단순히 텍스트만 저장
 2 => 쉼표 뿐만 아니라 탭 등으로 열 분리가 가능하여 혼란을 줄수 있음
 3 => 행/렬 표현으로 구조화된 데이터 표현 가능
 4= > csv는 xlsx과 비슷한 형태로 스프레드 시트에 열수 있음

### json이 웹개발에 사용 되는 이유 ?
 - 개발에서 널리 사용되는 이유는 단순히 구조를 명확하게 표현할 수 있다는 점뿐만 아니라, 간결성과 가독성, 처리 속도, 그리고 웹 환경과의 자연스러운 호환성 때문

1 => json 사람이 읽을 수 형태있는 텍스트 기반 데이터 포맷으로 바이너리 형태가 적합하지 않음
* 바이너리 형태 : 2진법과 같은 컴퓨터 언어
2 => 간결하고 읽기 쉬워 웹개발에 많이 사용
3 => txt가 가장 가벼움
4 => 사용하는 프로그래밍 언어나 라이브러리에 따라 차이가 있을 수 있음 (운영체제별 줄바꿈 방식 또는, 인코등 차이가 있을 수 있음)
